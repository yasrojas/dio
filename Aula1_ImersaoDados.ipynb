```python
import matplotlib.pyplot as plt

dados['tratamento'].value_counts().plot.pie(autopct='%1.1f%%', shadow= True)

plt.axis('equal')                                      
plt.title('Gráfico do tratamento')
plt.show()
```

```python
import matplotlib.pyplot as plt

dados['tempo'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Gráfico do tempo')
plt.axis('equal')
plt.show()
```
